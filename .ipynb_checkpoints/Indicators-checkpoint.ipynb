### RSI Formula:
$$
RSI_{step\,one} = 100 - [\frac{100}{1 + \frac{Average\,gain}{Average\,loss}}]\\
RSI_{step\,two} = 100 - [\frac{100}{1 + \frac{Previous\,average\,gain*13+Current\,gain}{Previous\,average\,loss*13+Current\,loss}}]
$$
که سود و زیان بصورت درصد تغییرات قیمت نسبت به دوره گذشته و هر دو با مقادیر مثبت نشان داده  میشوند، بنابراین سود متوسط برابر با متوسط سود های بدست آمده در بازه زمانی مورد نظر است که عموما چهارده روز در نظر گرفته میشود.

```python
'''
for all of this functions:
variabels:
    Inputs :    
        ohlcv : is a DataFrame with columns named "open, high, low, close, volume" with lowercase letters.
    outputs:
        the indicators
'''



def RSI(ohlcv, periods=14):
    n = periods
    prices = ohlcv['close']
    deltas = np.diff(prices)
    
    seed = deltas[:n+1]   
    
    up = seed[seed >= 0].sum()/n
    down = -seed[seed < 0].sum()/n
    rs = up/down
    rsi = np.zeros_like(prices)
    rsi[:n] = 100. - 100./(1.+rs)

    for i in range(n, len(prices)):
        delta = deltas[i-1]  # to get the i_th delta, since its index starts from zero 
        #delta = deltas.iloc[i-1]

        if delta > 0:
            upval = delta
            downval = 0.
        else:
            upval = 0.
            downval = -delta

        up = (up*(n-1) + upval)/n
        down = (down*(n-1) + downval)/n

        rs = up/down
        rsi[i] = 100. - 100./(1.+rs)

    return rsi

def SMA(values, window):
    '''
    Simple moving average with smooth starting
    '''
    values = np.asarray(values)
    smas = np.copy(values)
    # for first values in range of window, average is their own moving average, 
    # with growing window from one to window size
    for i in range(1,window):        
        smas[i] = (values[i] + smas[i-1]*i)/(i+1)
        
    weigths = np.repeat(1.0, window)/window
    smas[window-1:] = np.convolve(values, weigths, 'valid')
    
    return smas  # as a numpy array

def EMA(values, window):
    '''
    Exponential moving average
    '''
    weights = np.exp(np.linspace(-1., 0., window))
    weights /= weights.sum()
    a = np.convolve(values, weights, mode='full')[:len(values)]
    a[:window] = a[window]
    return a


def computeMACD(ohlcv, slow=26, fast=12):
    """
    compute the MACD (Moving Average Convergence/Divergence) using a fast and slow exponential moving avg'
    return value is emaslow, emafast, macd which are len(x) arrays
    """
    x = olhcv['close']
    emaslow = ExpMovingAverage(x, slow)
    emafast = ExpMovingAverage(x, fast)
    return emaslow, emafast, emafast - emaslow

def WILLIAMS(ohlcv, period = 14):
    
    highest_high = ohlcv["high"].rolling(center=False, window=period).max()
    lowest_low = ohlcv["low"].rolling(center=False, window=period).min()

    WR = pd.Series(
        (highest_high - ohlc["close"]) / (highest_high - lowest_low), 
        name=f"{period} Williams %R")

    return WR * -100
```
